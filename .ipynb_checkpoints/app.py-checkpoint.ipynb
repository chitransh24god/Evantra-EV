````python
from flask import Flask, request, jsonify, render_template
import google.generativeai as genai
import base64
import json
import re
import os
import mammoth
import tempfile

app = Flask(__name__)

# ── Configure Gemini ──────────────────────────────────────────────────────────
genai.configure(api_key="AIzaSyBgd2B5356PywjtjBCQIGzDKLFXr1dqSvw")
model = genai.GenerativeModel("gemini-1.5-flash")

# ── Helpers ───────────────────────────────────────────────────────────────────
def extract_json(text):
    text = text.strip()
    text = re.sub(r"```json|```", "", text).strip()
    match = re.search(r"\{.*\}", text, re.DOTALL)
    if match:
        return json.loads(match.group())
    raise ValueError("No JSON found in response")

def build_prompt(job_desc=""):
    jd_section = f"\n\nJOB DESCRIPTION:\n{job_desc[:1500]}" if job_desc else ""
    return f"""You are an expert ATS resume analyser and career coach.
Analyse the resume provided{' against the job description' if job_desc else ''}.{jd_section}

Return ONLY valid JSON (no markdown, no explanation):
{{
  "ats_score": <integer 0-100>,
  "score_breakdown": {{
    "formatting": <integer 0-100>,
    "keywords": <integer 0-100>,
    "experience": <integer 0-100>,
    "readability": <integer 0-100>
  }},
  "strengths": ["<3-5 short strings>"],
  "missing_keywords": ["<5-10 keyword strings>"],
  "skill_gaps": [{{"skill": "<name>", "priority": "high|medium|low"}}],
  "improvements": [{{"area": "<title>", "suggestion": "<1-2 sentence actionable advice>"}}],
  "verdict": "<2-3 sentence overall summary>"
}}"""

# ── Routes ────────────────────────────────────────────────────────────────────
@app.route("/")
def index():
    return render_template("index.html")

@app.route("/analyse", methods=["POST"])
def analyse():
    try:
        job_desc   = request.form.get("job_desc", "")
        pasted     = request.form.get("pasted_text", "")
        file       = request.files.get("resume_file")
        prompt_str = build_prompt(job_desc)

        # ── No file: use pasted text ──────────────────────────────────────────
        if not file or file.filename == "":
            if not pasted.strip():
                return jsonify({"error": "Please upload a file or paste resume text."}), 400
            full_prompt = prompt_str + f"\n\nRESUME TEXT:\n{pasted[:5000]}"
            response    = model.generate_content(full_prompt)
            result      = extract_json(response.text)
            return jsonify(result)

        filename = file.filename.lower()
        ext      = filename.rsplit(".", 1)[-1] if "." in filename else ""

        # ── Plain text ────────────────────────────────────────────────────────
        if ext == "txt":
            text         = file.read().decode("utf-8", errors="ignore")
            full_prompt  = prompt_str + f"\n\nRESUME TEXT:\n{text[:5000]}"
            response     = model.generate_content(full_prompt)
            result       = extract_json(response.text)
            return jsonify(result)

        # ── Word DOCX ─────────────────────────────────────────────────────────
        if ext in ("doc", "docx"):
            file_bytes = file.read()
            with tempfile.NamedTemporaryFile(suffix=".docx", delete=False) as tmp:
                tmp.write(file_bytes)
                tmp_path = tmp.name
            with open(tmp_path, "rb") as f:
                result_mammoth = mammoth.extract_raw_text(f)
            os.unlink(tmp_path)
            text        = result_mammoth.value
            if not text.strip():
                return jsonify({"error": "Could not extract text from this Word file."}), 400
            full_prompt = prompt_str + f"\n\nRESUME TEXT:\n{text[:5000]}"
            response    = model.generate_content(full_prompt)
            result      = extract_json(response.text)
            return jsonify(result)

        # ── PDF ───────────────────────────────────────────────────────────────
        if ext == "pdf":
            file_bytes  = file.read()
            b64_data    = base64.b64encode(file_bytes).decode("utf-8")
            response    = model.generate_content([
                {"mime_type": "application/pdf", "data": b64_data},
                prompt_str
            ])
            result      = extract_json(response.text)
            return jsonify(result)

        # ── Images (JPG, PNG, WEBP) ───────────────────────────────────────────
        if ext in ("jpg", "jpeg", "png", "webp", "gif"):
            mime_map    = {"jpg":"image/jpeg","jpeg":"image/jpeg",
                           "png":"image/png","webp":"image/webp","gif":"image/gif"}
            mime_type   = mime_map.get(ext, "image/png")
            file_bytes  = file.read()
            b64_data    = base64.b64encode(file_bytes).decode("utf-8")
            response    = model.generate_content([
                {"mime_type": mime_type, "data": b64_data},
                prompt_str
            ])
            result      = extract_json(response.text)
            return jsonify(result)

        return jsonify({"error": f"Unsupported file type: .{ext}"}), 400

    except json.JSONDecodeError:
        return jsonify({"error": "AI returned unexpected format. Please try again."}), 500
    except Exception as e:
        return jsonify({"error": str(e)}), 500

if __name__ == "__main__":
    app.run(debug=True, port=5000)
````
